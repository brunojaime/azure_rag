# Execute .sh file path

```python
import subprocess

def execute_sh(file_path):
    # Replace 'script.sh' with the path to your script
    result = subprocess.run([f'./{file_path}'], shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)


    # This will print the output and error (if any)
    print("STDOUT:", result.stdout.decode())
    print("STDERR:", result.stderr.decode())

    # Checking the exit code of the script (0 usually means successful execution)
    if result.returncode == 0:
        print("Script executed successfully.")
    else:
        print("Script failed with return code:", result.returncode)

```
